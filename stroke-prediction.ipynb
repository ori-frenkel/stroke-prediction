# Data Preparation

**We need to preform the following things:**
* Convert categorial features to numeric
* Handling missing values - the NaN values in the *bmi* feature & 'Unknown' *smoking status*
* Droping abnormal values - *other* value in the *gender* feature 
* Select our main features
* Solve the non-balanced data problem
* Split to train & test

Let's start preparing our data step by step:

## Convert Categorial Features to Numeric

```python
categorial_features = ['ever_married', 'Residence_type']

le = preprocessing.LabelEncoder()
for feature in categorial_features:
    stroke_data[feature] = le.fit_transform(stroke_data[feature])
    
stroke_data['gender'] = stroke_data['gender'].replace(['Female', 'Male', 'Other'], [0, 1, 2])
stroke_data['smoking_status'] = stroke_data['smoking_status'].replace(['never smoked', 'smokes', 'formerly smoked', 'Unknown'], [0, 1, 2, 3])
stroke_data['work_type'] = stroke_data['work_type'].replace(['Private', 'Govt_job', 'Self-employed', 'children', 'Never_worked'], [0, 1, 2, 3, 4])

stroke_data.head()
```

## Handling Missing Values

Replacing the NaN missing values in the *bmi* feature by taking the mean/Median of the other values:

```python
# Replace using median 
bmi_median = stroke_data['bmi'].median()
stroke_data['bmi'] = stroke_data['bmi'].fillna(bmi_median)
stroke_data.head()
```

Filling the 'Unkonwn' missing values in the *smoking_status* of chlidren with NEVER_SMOKED 

```python
smoking_status_counts = stroke_data['smoking_status'].value_counts()
unknown_count = smoking_status_counts.get(3)
print(f"Count of 'Unknown' smoking status before: {unknown_count}")

stroke_data.loc[(stroke_data['smoking_status'] == 3) & (stroke_data['work_type'] == 3), 'smoking_status'] = 0

smoking_status_counts = stroke_data['smoking_status'].value_counts()
unknown_count = smoking_status_counts.get(3)
print(f"Count of 'Unknown' smoking status after: {unknown_count}")
```

Predicting the 'Unkonwn' missing values in the *smoking_status* column using Logistic Regression:
1. We will split our data to train and test. **Training set** - all the records that do not have missing values in the *smoking_status* column. **Test set** - all the records that do have missing values in the *smoking_status* column.
1. Create the Logistic Regression and train it.
1. Use the trained model to complete the missing values.

```python
train = stroke_data[stroke_data.smoking_status != 3]
test = stroke_data[stroke_data.smoking_status == 3]

y_train = train['smoking_status']
X_train = train.drop(['id', 'smoking_status'],axis='columns')

X_test = test.drop(['id', 'smoking_status'],axis='columns')
```

```python
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
```

```python
smoking_status_preds = model.predict(X_test)
stroke_data.loc[stroke_data.smoking_status == 3, "smoking_status"] = smoking_status_preds
stroke_data.head()
```

## Droping Abnormal Values

To avoid bias, we will remove all records with *other* gender value. There are very few records like that and we will not want them to affect badly on our model.

```python
stroke_data[stroke_data.gender == 2]
```

```python
stroke_data = stroke_data[stroke_data.gender != 2]
stroke_data[stroke_data.gender == 2]
```

The outliers in *bmi* feature will be remove in the next section since we will drop the entire feature.

## Select Main Features

Let's look on the features heatmap correlation before we decide on which features we give up:

```python
plt.figure(figsize=(10,10))

sns.heatmap(stroke_data.corr(), annot=True, fmt='.2f', cmap=sns.color_palette("YlOrBr", as_cmap=True))

plt.title("Features Heatmap", weight='bold', fontsize=18)
plt.show()
```

**We will drop the following features** - [*id, Residence_type, gender, bmi*]

1. *id* - Of course not affecting the decision if a person has stroke or not
1. *Residence_type* - we saw earler in the 'Distribution of Stroke by Residence' diagram that stroke occurs in urban and also in rural almost at the same probability. In addition, we can see that correlation between this feature and stroke is very low (-0.02). All of these lead us to the decision that this feature will not help us determine whether person will experience stroke.
1. *gender* - we saw earler that stroke occurs for male and female almost at the same probability. In addition, we can see that correlation between this feature and stroke is very low (0.01). Additionally, it has low correlation to all other features. All of these lead us to the decision that this feature will not help us determine whether person will experience stroke.
1. *bmi* - while BMI does have a slightly stronger correlation with certain variables, its overall relationship with the target feature, stroke, is weak. Hence, it was deemed unnecessary for model prediction. (We tried as well with bmi, and the model was weaker)

```python
stroke_data = stroke_data.drop(['id', 'Residence_type', 'bmi', 'gender'], axis='columns')
stroke_data.head()
```

removing the target feature:

```python
target = stroke_data['stroke']
stroke_data = stroke_data.drop('stroke',axis='columns')
stroke_data.head()
```

## Split to Train & Test

```python
# split our data to train and test groups
X_train, X_test, y_train, y_test = train_test_split(stroke_data, target, test_size=0.25, random_state=42)
```

## Scaling

```python
# Normalize numerical features
numerical_features = ['age', 'avg_glucose_level']

scaler = StandardScaler()
X_train[numerical_features] = scaler.fit_transform(X_train[numerical_features])
X_test[numerical_features] = scaler.transform(X_test[numerical_features])
```

## Data Balancing

In order to fix this problem, we have to generate more samples of stroke records. To do so, we will use oversampling & undersampling techniques. These techniques will only be applied to the training set to prevent data leakage.

```python
# the data is not balance
print(Counter(y_train))

# balance the data - generate more stroke sampels
over = SMOTE(sampling_strategy=0.5)
under = RandomUnderSampler(sampling_strategy=0.8)
pipeline = Pipeline([('over', over), ('under', under)])
X_train, y_train = pipeline.fit_resample(X_train, y_train)

# now the data is balance
print(Counter(y_train))
```

# Modeling

We will examine several different models:

```python
models = [DummyClassifier(strategy='most_frequent'), KNeighborsClassifier(), DecisionTreeClassifier(), 
          RandomForestClassifier(), GradientBoostingClassifier(), XGBClassifier(eval_metric='mlogloss')]
models_names = ['DummyClassifier', 'KNeighborsClassifier', 'DecisionTreeClassifier', 
                'RandomForestClassifier', 'GradientBoostingClassifier', 'XGBClassifier']
```

```python
def model_score(model, X, Y):
    accuracy_value = cross_val_score(model, X, Y, scoring='accuracy', cv=5).mean()
    recall_value = cross_val_score(model, X, Y, scoring='recall', cv=5).mean()
    precision_value = cross_val_score(model, X, Y, scoring='precision', cv=5).mean()
    f1_value = cross_val_score(model, X, Y, scoring='f1', cv=5).mean()
    return accuracy_value, recall_value, precision_value, f1_value
```

```python
accuracy,recall,precision,f1 = [],[],[],[]

for model in models:
    accuracy_value, recall_value, precision_value, f1_value = model_score(model, X_train, y_train)
    accuracy.append(accuracy_value)
    recall.append(recall_value)    
    precision.append(precision_value)    
    f1.append(f1_value)    
```

```python
models_performence = pd.DataFrame({'Model': models_names, 'Accuracy': accuracy, 'Recall': recall, 'Precision': precision, 'F1': f1})
models_performence.style.background_gradient(cmap=sns.color_palette("YlOrBr", as_cmap=True), high=1)
```

The model that we will choose is **RandomForest**. 

Although KNN achieves the highest recall, which aligns with our primary focus, we ultimately selected the Random Forest model due to its balanced performance across multiple metrics. This balance ensures that we prioritize recall without compromising the overall quality and usability of the model.

Now let's try improve model preformence by choosing a set of optimal hyperparameters for a learning algorithm.

## Hyperparameter Tuning

Let's find the best hyperparameters for the RandomForest model.

We will use the Grid Search technique for finding the best hyperparameters values from finite set we defined below.

```python
random_forest = RandomForestClassifier()
hyperparameter_values = {'n_estimators': [100, 200, 300],
                          'max_depth': [10, 20, None],
                          'max_features': ['sqrt', 'log2', None],
                          'bootstrap': [True, False]
                        }

kfold_3 = KFold(n_splits=3, shuffle=True)
clf = GridSearchCV(random_forest, param_grid=hyperparameter_values, cv=kfold_3, scoring=make_scorer(recall_score))

# Fit the grid search to the data
clf.fit(X_train, y_train)
random_forest_best_params = clf.best_params_

random_forest_best_params
```

We will build the Random Forest model with the best hyperparameters and compare it to the previous results:

```python
random_forest = RandomForestClassifier(**random_forest_best_params)

accuracy_value, recall_value, precision_value, f1_value = model_score(random_forest, X_train, y_train)
best_scoring = [recall_value]
prev_scoring = [recall[3]]
metrics = ['Recall']

models_performence = pd.DataFrame({'Metrics': metrics, 'Prev Random Forest': prev_scoring, 'Best Random Forest': best_scoring})
```

```python
ax = models_performence.plot(kind="bar", x='Metrics', y=['Prev Random Forest', 'Best Random Forest'],
                            color=sns.color_palette("YlOrBr"), figsize = (8, 6))
ax.set(ylim=[0.8, 0.95])
for p in ax.patches:
    b = p.get_bbox()
    ax.annotate("{:.3f}".format(b.y1 + b.y0), (b.x0, b.y1+0.001), fontsize=9.5, weight='bold')
```

As you can see, we successfully improved our most important metric - recall.

Our final model will be Random Forest with this hyperparameters values.
Let's build and train the model:

```python
model = RandomForestClassifier(**random_forest_best_params)
model.fit(X_train, y_train)
```

# Evaluation

We will evaluate our model using several methods:

## Confusion Matrix

```python
pred = model.predict(X_test)
conf_matrix = confusion_matrix(y_test, pred)

# Print the confusion matrix using Matplotlib
fig, ax = plot_confusion_matrix(conf_mat=conf_matrix, figsize=(6, 6), cmap=sns.color_palette("YlOrBr", as_cmap=True))
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()

print("False Positive Rate: {:.3f}".format(conf_matrix[0][1] / (conf_matrix[0][0] + conf_matrix[0][1])))
print("False Negative Rate: {:.3f}".format(conf_matrix[1][0] / (conf_matrix[1][0] + conf_matrix[1][1])))
```

## Classification Report

```python
print(classification_report(y_test, pred))
```
